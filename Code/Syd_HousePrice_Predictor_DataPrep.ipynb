## to load demographis data for suburbs

```python
# DO NOT RUN THIS CODE SINCE THIS MIGHT EXHAUST DAILY API CALLS#JSON  FILE SAVED IN RESOURCES FOLDER
def get_property_info(suburb,postcode):
    response = requests.post(auth_url,
        data = {
        'client_id':client_id,
        'client_secret':client_secret,
        'grant_type':'client_credentials',
        'scope': scopes_2,
        'Content-Type':'text/json'}
        )
    json_res = response.json()
    access_token=json_res["access_token"]
    print(access_token)
    auth = {'Authorization':'Bearer '+access_token}
    

    url_detailed_de='https://api.domain.com.au/v2/demographics/NSW/'+suburb+'/'+str(postcode)+'?year=2016'
    print(url_detailed_de)
    res_de = requests.get(url_detailed_de,headers = auth)
    r_de=res_de.json()
    json_object =json.dumps(r_de,indent=4)
    json.dumps(r_de,indent=4)
            

    

    with open("./Resources/{}_{}.json".format(suburb,postcode),"w") as outfile:
            outfile.write(json_object)   

```

```python
for index, row in list_suburbs.iterrows():   
    try:
        get_property_info(row['suburb'],row['postcode'])
        print('Success', row['suburb'],row['postcode'])
        final_list_suburbs.append(row['suburb'])

    except Exception as e:
       
        print('Failure in',row['postcode'],row['suburb'],e)
        pass
```

```python
# loop for list suburbs to cal and  extract unemployment rate 


import traceback
df_demo_frame=[]
df_demo_unemp = pd.DataFrame()

for index, row in list_suburbs.iterrows(): 
    suburb=row['suburb']
    postcode=row['postcode']
    try:
        with open("./Resources/{}_{}.json".format(suburb,postcode),"r") as sub_demo_data:
            sub_demo_data=json.load(sub_demo_data)
            for i in [0]:

                # to cal unemployment rate in suburb selected
                df_demo = pd.DataFrame(sub_demo_data['demographics'][i]['items'])
                df_demo['Un_rate_%'] = 100*df_demo['value'] / df_demo['value'].sum()
                df_demo[['suburb','postcode']]=suburb,postcode


        df_demo_frame.append(df_demo.loc[df_demo['label']=='Unemployed','Un_rate_%':'postcode'])
        df_demo_unemp=pd.concat(df_demo_frame)
        
    except KeyError as e:
        print(traceback.format_exc(),suburb)
```
